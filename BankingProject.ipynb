```python
import pandas as pd
import numpy as np
import pymysql
from sqlalchemy import create_engine
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#mysql
#host = localhost
#port = 3306
#user = root
#password = 'root'
#mysql connection
engine_mysql = create_engine("mysql+pymysql://root:root@localhost:3306/walmart_db")

try:
    engine_mysql
    print("Connection Sucessful")
except:
    print("Unable to connect")
```

```python
query = "Select * from banking_data.customer"
```

```python
data = pd.read_sql(query, engine_mysql)
```

```python
bins = [0, 100000, 300000, float('inf')]
labels = ['Low', 'Mid', 'High']
data['Income Band'] = pd.cut(data['Estimated Income'], bins=bins, labels=labels, include_lowest=True)
```

```python
categorical_cols = data[["BRId", "GenderId", "IAId", "Amount of Credit Cards","Risk Weighting","Nationality","Occupation","Fee Structure","Loyalty Classification","Properties Owned","Income Band"]].columns

for col in categorical_cols:
  print(f"Value Counts for '{col}':")
  display(data[col].value_counts())
```

```python
null_values = data.isnull().sum()
print("Missing values per column:\n", null_values)
```

```python

cols = ['Nationality','Amount of Credit Cards','Fee Structure','Estimated Income', 'Superannuation Savings', 'Credit Card Balance', 'Bank Loans', 'Bank Deposits', 'Checking Accounts', 'Saving Accounts', 'Foreign Currency Account', 'Business Lending']

# Univariate analysis and visualization
plt.figure(figsize=(8,50))
for i, col in enumerate(cols):
    plt.subplot(12, 1, i + 1)
    sns.histplot(data[col], kde=True)
    plt.title(col)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=data, x='Amount of Credit Cards', hue='GenderId')
```

```python
for i, pred in enumerate(data[["BRId", "GenderId", "IAId",'Amount of Credit Cards','Fee Structure','Income Band']].columns):
    plt.figure(i)
    sns.countplot(data=data, x=pred, hue='Nationality')
```

```python

numerical_cols = ['Age', 'Estimated Income', 'Superannuation Savings', 'Credit Card Balance',
                  'Bank Loans', 'Bank Deposits', 'Checking Accounts', 'Saving Accounts',
                  'Foreign Currency Account', 'Business Lending', 'Properties Owned']


correlation_matrix = data[numerical_cols].corr()


plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='crest', fmt=".2f")
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```
